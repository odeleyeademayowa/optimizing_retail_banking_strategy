```python
# Define the datetime objects for 1900 and 2016
start_date = pd.to_datetime('1900-01-01')
end_date = pd.to_datetime('2016-10-21')

# Filter for extreme DOBs
extreme_dob_df = bank_df[(bank_df['CustomerDOB'] < start_date) | (bank_df['CustomerDOB'] > end_date)]

# View the first few extreme DOBs
print(extreme_dob_df[['CustomerID', 'CustomerDOB']].head())

# Count of extreme cases
print("\nNumber of customers with extreme Date of Birth:", extreme_dob_df.shape[0])
```

```python
# Replace extreme DOBs (before 1900 or after 2016) with the median DOB
valid_dob_median = bank_df[(bank_df['CustomerDOB'] >= start_date) & (bank_df['CustomerDOB'] <= end_date)]['CustomerDOB'].median()

# Replace invalid DOBs with the median DOB
bank_df.loc[(bank_df['CustomerDOB'] < start_date) | (bank_df['CustomerDOB'] > end_date), 'CustomerDOB'] = valid_dob_median
```

```python
# Check if replacement was successful
print(bank_df[bank_df['CustomerDOB'] < start_date].shape[0])  
print(bank_df[bank_df['CustomerDOB'] > end_date].shape[0])  
```

```python
# Define segmentation function
def age_group(age):
    if 16 <= age <= 25:
        return 'Youth'
    elif 26 <= age <= 35:
        return 'Young Adult'
    elif 36 <= age <= 45:
        return 'Adult'
    elif 46 <= age <= 55:
        return 'Mature Adult'
    elif 56 <= age <= 65:
        return 'Elder'
    elif 66 <= age <= 75:
        return 'Senior'
    elif 76 <= age <= 100:
        return 'Aged'
    else:
        return 'Unknown'

# Apply the segmentation
bank_df['AgeSegment'] = bank_df['Age'].apply(age_group)

# View distribution
print(bank_df['AgeSegment'].value_counts())
```

# RFM ANALYSIS

```python
# Recency: Days since last transaction
latest_date = bank_df['TransactionDate'].max()
recency = bank_df.groupby(['CustomerID']).agg({'TransactionDate': lambda x: ((latest_date - x.max()).days) +1}).rename(columns={'TransactionDate': 'Recency'})
```

```python
recency.head()
```

```python
# Frequency: Number of transactions per customer
frequency = bank_df.drop_duplicates(subset = 'TransactionID').groupby(['CustomerID'])[['TransactionDate']].count().rename(columns={'TransactionDate': 'Frequency'})
```

```python
frequency.head()
```

```python
monetary = bank_df.groupby('CustomerID')[['TransactionAmount']].sum().rename(columns={'TransactionAmount': 'Monetary'})
```

```python
monetary.head()
```

```python
# RFM Table
rfm = recency.merge(frequency, on='CustomerID')
rfm = rfm.merge(monetary, on='CustomerID')
```

```python
# Plot histograms
plt.figure(figsize=(20, 6))

# Recency
plt.subplot(1, 3, 1)
plt.hist(rfm['Recency'], bins=10, color='skyblue', edgecolor='black')
plt.title('Recency Distribution')
plt.xlabel('Days Since Last Transaction')
plt.ylabel('Number of Customers')

# Frequency
plt.subplot(1, 3, 2)
plt.hist(rfm['Frequency'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Frequency Distribution')
plt.xlabel('Number of Transactions')

# Monetary
plt.subplot(1, 3, 3)
plt.hist(rfm['Monetary'], bins=10, color='salmon', edgecolor='black')
plt.title('Monetary Distribution')
plt.xlabel('Total Spend')

plt.tight_layout()
plt.show()
```

### Recency - Most customers had their last transaction between 4080 days ago. Very few customers are recent  this suggests low engagement or infrequent activity across the customer base.
### Frequency - Most customers had only one transaction. The frequency drops off steeply after that: very few customers make more than 23 transactions and even fewer more than 3 or 4 times.
### - Business Insight: This long tail of low-frequency customers, indicates a need for re-engagement strategies.
### Monetary - Almost all customers are clustered at very low total spend (< 0.2 million).
### - There's extreme skewness, possibly a few high-value customers skew the scale

```python
# Account Balance Tier
def balance_tier(balance):
    if balance < 1000:
        return 'Very Low'
    elif balance < 10000:
        return 'Low'
    elif balance < 100000:
        return 'Moderate'
    elif balance < 1000000:
        return 'High'
    elif balance < 10000000:
        return 'Very High'
    else:
        return 'Ultra High'

bank_df['BalanceTier'] = bank_df['CustAccountBalance'].apply(balance_tier)
```

# Assigning scores to the rfm metric

```python
# Calculate quantiles
r_quartiles = rfm['Recency'].quantile([0.25, 0.5, 0.75])
f_quartiles = rfm['Frequency'].quantile([0.25, 0.5, 0.75])
m_quartiles = rfm['Monetary'].quantile([0.25, 0.5, 0.75])
```

```python
r_quartiles
```

```python
f_quartiles
```

```python
m_quartiles
```

```python
# checking frequency counts
rfm['Frequency'].value_counts()
```

```python
# Scoring functions
def r_score(x):
    if x <= r_quartiles[0.25]:
        return 4
    elif x <= r_quartiles[0.5]:
        return 3
    elif x <= r_quartiles[0.75]:
        return 2
    else:
        return 1

def custom_f_score(x):
    if x <= 3:
        return x
    else:
        return 4

def m_score(x):
    if x <= m_quartiles[0.25]:
        return 1
    elif x <= m_quartiles[0.5]:
        return 2
    elif x <= m_quartiles[0.75]:
        return 3
    else:
        return 4
```

```python
rfm['RecencyScore'] = rfm['Recency'].apply(r_score)
rfm['FrequencyScore'] = rfm['Frequency'].apply(custom_f_score)
rfm['MonetaryScore'] = rfm['Monetary'].apply(m_score)
```

```python
rfm
```

```python
rfm['RFM_Score'] = rfm[['RecencyScore', 'FrequencyScore', 'MonetaryScore']].sum(axis = 1)
```

```python
rfm['RFM_Group'] = rfm['RecencyScore'].astype(str) + \
                  rfm['FrequencyScore'].astype(str) + \
                  rfm['MonetaryScore'].astype(str)
```

```python
def segment_customer(score):
    if score >= 10:
        return 'High-Value'
    elif score >= 5:
        return 'Loyal'
    elif score >= 4:
        return 'At-Risk'
    else:
        return 'Low-Value'

rfm['CustomerSegment'] = rfm['RFM_Score'].apply(segment_customer)
```

# Unsupervised Learning using K-Means Clustering 

```python
import sklearn
from sklearn.cluster import KMeans

!pip install yellowbrick
import yellowbrick
from yellowbrick.cluster import KElbowVisualizer
from sklearn.decomposition import PCA
```

```python
rfm_cluster = rfm.drop(['AvgTransactionValue', 'RFM_Group', 'CustomerSegment', 'Weighted_RFM_Score',	'Weighted_CustSegment',	'MarketingStrategy'], axis=1)
rfm_cluster.head()
```

```python
# Set 'CustomerID' as the index
rfm_cluster.set_index('CustomerID', inplace=True)
```

```python
# Scale the features for KMeans clustering
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_cluster)
```

```python
# Using K-Means Clustering to verify or enhance your segments based on RFM values.
# Initialize KMeans model
model_1 = KMeans(random_state=42)

# Use Yellowbrick to visualize the elbow method
visualizer = KElbowVisualizer(model_1, k=(1, 10))  # Test clusters from 1 to 10
visualizer.fit(rfm_scaled)  # Fit the model
visualizer.show()  # Display the elbow plot
```

```python
model_2 = KMeans(random_state=42)
# Create ElbowVisualizer using Calinski-Harabasz metric
visualizer = KElbowVisualizer(model_2, k=(2, 11), metric='calinski_harabasz', locate_elbow=True, timings=False)

# Fit and show
visualizer.fit(rfm_scaled)
visualizer.show()
```

```python
# Fit KMeans with optimal number of clusters 
optimal_k = 3   

final_model = KMeans(n_clusters=optimal_k, random_state=42)
rfm['Cluster'] = final_model.fit_predict(rfm_scaled)

# Print the mean values of each cluster to better understand them
print(rfm.groupby('Cluster')[['Recency', 'Frequency', 'Monetary', 'RecencyScore', 'FrequencyScore',	'MonetaryScore']].mean())
```
